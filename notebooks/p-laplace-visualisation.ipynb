# Vizualizace odhadu 1. vlastního čísla p-laplacianu 

```julia
using DrWatson
@quickactivate "Eigenvalue bounds of p-Laplacian"

using Plots; gr()

include(srcdir("pLaplaceNumerics.jl"));
```

```julia
n = 31 # number of division points
ns = 1 # number of subdivision points of each subinterval

λ₁ᵉˣᵃᶜᵗ(P) = (P-1)*(2*(π/P)/(sin(π/P)))^P

p = 2 
λ₁ = λ₁ᵉˣᵃᶜᵗ(p)
λᵢₙᵢₜ = (3.,1.5*λ₁)

t, tᴵ, U₁, U₁ᴵ, U₂, U₂ᴵ, U₁d, U₁dᴵ, Λ₁ = plaplace_solve(λᵢₙᵢₜ, p, n); #, 
U₁[end]=0.


ΔₚU_tmp = abs.(U₁d).^(p-2).*U₁d
ΔₚU = [0.,]
for i in 1:length(ΔₚU_tmp)-1
    #append!(U₂d, (U₂[i+1]-U₂[i])/(t[i+1]-t[i]))
    append!(ΔₚU, (ΔₚU_tmp[i+1]-ΔₚU_tmp[i])/(t[i+1]-t[i]))
end


ΔₚU_zU2 = [0.,]
for i in 1:length(U₂)-1
    append!(ΔₚU_zU2, (U₂[i+1]-U₂[i])/(t[i+1]-t[i]))
#    append!(ΔₚU, (ΔₚU_tmp[i+1]-ΔₚU_tmp[i])/(t[i+1]-t[i]))
end

p1 = plot(t,U₁,label="U₁", marker=3)
p2 = plot(t,U₂,label="U₂", marker=3)
#p3 = plot(t,(-U₂d./U₁.^(p-1)), label="ΔₚU", marker=3)
p3 = plot(t,ΔₚU, label="ΔₚU z U₁'", marker=3)
p4 = plot(t,ΔₚU_zU2, label="ΔₚU z U₂", marker=3)
#p4 = plot(t,U₁d,label="U₁ derivace", marker=3)
plot(p1,p2,p3, p4, layout=(2,2), size=(1200,800))
```
